# **QAI for SysEngg Demo: Drone Subsystem**

QAI for Systems Engineering framework inside Bhadale IT Org context. It clearly shows:

**Client Needs Intake**
→ AeroTech’s constraints/requirements are captured.

**System Decomposition**
→ Navigation, Control, Comms, Power broken down into subsystems.

**Framework Mapping**
→ Bhadale IT steps aligned with INCOSE, NASA, DoD, TOGAF milestones.

**Requirements Traceability** (DOORS/Jaama-like)
→ REQ IDs mapped to subsystems and solutions.

**TRL Progression**
→ Evolution from concept → operational.

**Final Product Snapshot**
→ Delivered subsystem with features + compliance references.

```python
# QAI for Systems Engineering Products Framework - Demo
# Example: Drone Subsystem Development
# Colab-ready code

import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

# ======================================================
# STEP 1: CLIENT INTAKE
# ======================================================
client_request = {
    "client": "AeroTech Pvt Ltd",
    "domain": "Aerospace & Defense",
    "need": "Autonomous Drone subsystem with QAI navigation + obstacle avoidance",
    "constraints": ["Weight < 2kg", "Flight time > 40 mins", "Secure comms"]
}

print("=== CLIENT INTAKE ===")
for k,v in client_request.items():
    print(f"{k}: {v}")

# ======================================================
# STEP 2: SYSTEM DECOMPOSITION
# ======================================================
subsystems = {
    "Navigation": ["QAI-based path planning", "Sensor fusion"],
    "Control": ["Flight stabilization", "Adaptive PID"],
    "Comms": ["Post-quantum secure channel", "Cloud interface"],
    "Power": ["Battery optimization", "Lightweight materials"]
}

print("\n=== SYSTEM DECOMPOSITION ===")
for s, fns in subsystems.items():
    print(f"{s}: {', '.join(fns)}")

# ======================================================
# STEP 3: FRAMEWORK MAPPING (INCOSE, NASA, DoD, TOGAF)
# ======================================================
framework_mapping = pd.DataFrame([
    ["Requirements Capture", "Stakeholder Needs", "Mission Needs Statement", "Capability Development Doc", "Architecture Vision"],
    ["Design", "Logical Architecture", "System Design Review (SDR)", "Preliminary Design Review (PDR)", "Business Architecture"],
    ["Integration & Test", "Verification/Validation", "System Integration Review (SIR)", "Test Readiness Review (TRR)", "Technology Architecture"],
    ["Deployment", "Transition Planning", "Operational Readiness Review (ORR)", "Production Readiness Review (PRR)", "Implementation Governance"]
],
    columns=["Bhadale IT Org", "INCOSE", "NASA", "DoD", "TOGAF"]
)

print("\n=== FRAMEWORK MAPPING ===")
print(framework_mapping)

# ======================================================
# STEP 4: REQUIREMENTS TRACEABILITY (simulated IBM DOORS/Jama output)
# ======================================================
traceability = pd.DataFrame([
    ["REQ-001", "Drone shall fly >40 min", "Power", "Linked to Battery Optimization"],
    ["REQ-002", "Drone shall avoid obstacles", "Navigation", "Linked to QAI Path Planning"],
    ["REQ-003", "Drone comms shall be secure", "Comms", "Linked to PQ-secure Channel"]
],
    columns=["Req ID", "Requirement", "Subsystem", "Trace Link"]
)

print("\n=== REQUIREMENTS TRACEABILITY ===")
print(traceability)

# ======================================================
# STEP 5: TRL PROGRESSION
# ======================================================
trl_evolution = pd.DataFrame({
    "Phase": ["Concept", "Prototype", "Subsystem Test", "System Demo", "Operational Use"],
    "TRL": [2, 4, 5, 7, 9]
})

print("\n=== TRL PROGRESSION ===")
print(trl_evolution)

# Plot TRL Progress
plt.figure(figsize=(6,4))
plt.plot(trl_evolution["Phase"], trl_evolution["TRL"], marker='o')
plt.title("TRL Evolution of Drone Subsystem")
plt.xlabel("Phase")
plt.ylabel("TRL Level")
plt.grid(True)
plt.show()

# ======================================================
# STEP 6: FINAL PRODUCT SNAPSHOT
# ======================================================
final_product = {
    "Product": "QAI-Enabled Drone Subsystem",
    "Delivered To": client_request["client"],
    "Features": [
        "Autonomous QAI navigation",
        "Adaptive flight control",
        "Post-quantum secure communication",
        "40+ min endurance"
    ],
    "Compliance": ["INCOSE SE Handbook", "NASA NPR 7123", "DoD 5000.02", "TOGAF ADM"]
}

print("\n=== FINAL PRODUCT SNAPSHOT ===")
for k,v in final_product.items():
    print(f"{k}: {v}")

# ======================================================
# STEP 7: VISUAL FLOW DIAGRAM
# ======================================================
G = nx.DiGraph()
steps = [
    "Client Intake", "System Decomposition", "Framework Mapping",
    "Traceability", "TRL Progression", "Final Product"
]

for i in range(len(steps)-1):
    G.add_edge(steps[i], steps[i+1])

plt.figure(figsize=(8,5))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=2500, font_size=9, font_weight="bold", arrows=True)
plt.title("QAI for Systems Engineering Framework Flow")
plt.show()
```
